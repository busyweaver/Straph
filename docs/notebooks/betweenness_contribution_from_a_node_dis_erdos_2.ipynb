```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (25,20)
import straph as sg
import straph.betweenness as bt
import straph.paths.meta_walks as mw
path_directory = "examples/"
S = sg.read_stream_graph(path_nodes=path_directory + "erdos_dis"+str(nb_erdos)+"_nodes.sg",
                      path_links=path_directory + "erdos_dis"+ str(nb_erdos)+ "_links.sg")
S.describe()
```

```python
import operator
import networkx as nx
b = operator.lt
walk_type = "passive"
fun =  mw.Metawalk.co_first_arrival
```

```python
G = bt.predecessor_graph_dis_gen(S, pre,node)
preced = bt.preced_node(S, G,events,events_reverse)
node_inf, clos_inf = bt.remove_infinite_from_predecessor_dis_gen(S, G, events, events_reverse,
                                                                 opt_walk, cur_best, fun, len(S.nodes),
                                                                 walk_type, b)
GG = bt.graph_to_ordered(G, events, events_reverse)
```

```python
import networkx as nx
GD = G.graph
ren = dict()
for e in GD.nodes:
    ren[e] = (S.node_to_label[e[0]],e[1])
#print(ren)
GN =  nx.relabel_nodes(GD, ren, copy=True)
fig, ax = plt.subplots(figsize=(30,30))
nx.draw_networkx(GN,arrows=True,width = 1,ax = ax)
```
