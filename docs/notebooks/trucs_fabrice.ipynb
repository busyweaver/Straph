```python
#k varie entre 0 et n-1
def s(n,m,k):
    return sum(   binomial(n-1,k+l)*(2*m/(n*(n-1)))^(k+l)*(1-2*m/(n*(n-1)))^(n-1-k-l)*binomial(k+l,k)*(1/2)^(k+l)  for l in range(0,n-1-k+1) )
```

```python
#k varie entre 0 et n-1
def s2(n,m,k):
    return binomial(n-1,k)*(m/(n*(n-1)))^k*(1-m/(n*(n-1)))^(n-1-k)
```

```python
for nn in range(1,10):
    for mm in range(1,binomial(nn,2)):
        for kk in range(1,n-1):
            if s(nn,mm,kk) != 0:
                print(s(nn,mm,kk), s2(nn,mm,kk) )
```
