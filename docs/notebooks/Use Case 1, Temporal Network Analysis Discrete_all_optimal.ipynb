```python
nb_erdos = 2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (25,20)
import straph as sg
import straph.betweenness as bt
import straph.paths.meta_walks as mw
path_directory = "examples/"
S = sg.read_stream_graph(path_nodes=path_directory + "erdos_dis"+str(nb_erdos)+"_nodes.sg",
                      path_links=path_directory + "erdos_dis"+ str(nb_erdos)+ "_links.sg")
S.describe()
```

```python
T = [0,10]
nb_node = 20
prob = 0.1
S = sg.erdos_renyi_discrete(T, nb_node,prob)
S.describe()
```

```python
bt.computation_betweenness_many_optimal_dis_gen(S, "erdos3", approx = -1)
```

```python
between = bt.read_dictionary("erdos3_active_foremost_exact_betweenness.pic")
```

```python
d = [ S.node_to_label[v]  for v in S.nodes ]
```

```python
ax = bt.heatmap_betweenness(S, between, d, square = True)
fig = ax.get_figure()
fig.savefig("out.png") 
```
